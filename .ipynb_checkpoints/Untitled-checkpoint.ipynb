```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [20, 10]
plt.rcParams["figure.dpi"] = 200
plt.style.use('seaborn-notebook')
import matplotlib as mpl
from datetime import datetime, timedelta, timezone
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

import ccxt
import exchange_poloniex
```

```python
mdf = exchange_poloniex.fetch_df()
```

```python
print(mdf.shape)
mdf.head()
```

```python
plt.plot(mdf.posix, mdf.open, 'b-', label='open')
plt.plot(mdf.posix, mdf.close, 'k-', label='close')
plt.legend(loc='best')
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(mdf.posix, mdf.open, 'b-')
ax1.set_xlabel('opening')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('open', color='b')
ax1.tick_params('y', colors='b')
ax1.grid(False)

ax2 = ax1.twinx()
ax2.plot(mdf.posix, mdf.close, 'k-')
ax2.set_ylabel('close', color='k')
ax2.tick_params('y', colors='k')
ax2.grid(False)

ax3 = ax2.twinx()
ax3.plot(mdf.posix, mdf.high, 'g-')
ax3.set_ylabel('high', color='g')
ax3.tick_params('y', colors='g')
ax3.grid(False)

ax4 = ax3.twinx()
ax4.plot(mdf.posix, mdf.low, 'r-')
ax4.set_ylabel('low', color='r')
ax4.tick_params('y', colors='r')
ax4.grid(False)

# fig.tight_layout()
plt.show()
```
