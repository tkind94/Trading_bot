```python
# convert series to supervised learning
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg
```

```python
# load dataset
dataset = expx.load_df()[['open']]
values = dataset.values

# ensure all data is float
values = values.astype('float32')

# normalize features
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)

# specify the number of lag hours
n_timesteps_in = 5
n_timesteps_out = 1

# frame as supervised learning
reframed = series_to_supervised(data = scaled, n_in = n_timesteps_in, n_out = n_timesteps_out)

print(reframed.shape)
reframed.head()
```

```python
# split into train and test sets
values = reframed.values
tts = int(len(values) * 0.7)
train = values[:tts, :]
test = values[tts:, :]
# split into input and outputs
n_features = int(reframed.shape[1]/(n_timesteps_in+1))
n_obs = n_timesteps_in * n_features
n_prd = reframed.shape[1] - n_obs
train_X, train_y = train[:, :n_obs], train[:, -n_prd:]
test_X, test_y = test[:, :n_obs], test[:, -n_prd:]
print(train_X.shape, len(train_X), train_y.shape)
# reshape input to be 3D [samples, timesteps, features]
train_X = train_X.reshape((train_X.shape[0], n_timesteps_in, n_features))
# train_y = train_y.reshape((train_y.shape[0], n_timesteps_out, n_features))
test_X = test_X.reshape((test_X.shape[0], n_timesteps_in, n_features))
# test_y = test_y.reshape((test_y.shape[0], n_timesteps_out, n_features))
print(train_X.shape, train_y.shape, test_X.shape, test_y.shape)
```

```python
# design network
model = Sequential()
model.add(Dense(64, input_shape=(train_X.shape[1], train_X.shape[2])))
# model.add(CuDNNLSTM(64, input_shape=(train_X.shape[1], train_X.shape[2]), return_sequences=True))
# model.add(CuDNNLSTM(64, return_sequences=True))  # returns a sequence of vectors of dimension 32
# model.add(CuDNNLSTM(64, return_sequences=True))  # returns a sequence of vectors of dimension 32
# model.add(CuDNNLSTM(64))  # return a single vector of dimension 32
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.001))
model.add(Dense(15, activation='relu'))
model.add(Dropout(0.001))
model.add(Dense(output_dim = train_y.shape[1], activation = 'linear')) #Output layer
model.compile(loss='mean_squared_error', optimizer='nadam', metrics=['mse', 'accuracy'])
```

```python
# fit network
history = model.fit(train_X, train_y, epochs=2000, batch_size=1024, validation_data=(test_X, test_y), verbose=1, shuffle=False)
```

```python
# plot training
view = 1000
f, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)

ax1.set_title('Loss')
ax1.plot(history.history['loss'][-view:], label='train')
ax1.plot(history.history['val_loss'][-view:], label='test')

ax2.set_title('Accuracy')
ax2.plot(history.history['acc'][-view:], label='train')
ax2.plot(history.history['val_acc'][-view:], label='test')

ax3.set_title('MSE')
ax3.plot(history.history['mean_squared_error'][-view:], label='train')
ax3.plot(history.history['val_mean_squared_error'][-view:], label='test')

plt.legend()
plt.show()
```

```python
model.evaluate(x=test_X, y=test_y)
```

```python
def predict(df, points):
    before = df[:, 0, -1].flatten()
    plt.plot(before, label='before')
    for point in points:
        after = model.predict(df[:point])[-1]
        padding = [None for p in range(point+1)]
        plt.plot(np.concatenate([padding, after]), label='after')
    plt.legend()
    plt.show() 
```
