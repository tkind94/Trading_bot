```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [30, 10]
plt.rcParams["figure.dpi"] = 250
plt.style.use('seaborn-notebook')
import matplotlib as mpl
from datetime import datetime, timedelta, timezone
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

import ccxt
import exchange_poloniex
```

```python
mdf = exchange_poloniex.fetch_df()
```

```python
print(mdf.shape)
mdf.head()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(mdf.posix, mdf.open, 'k-')
ax1.set_xlabel('posix')
ax1.set_ylabel('open', color='k')
ax1.grid(False)

ax2 = ax1
ax2.plot(mdf.posix, mdf.close, 'b-')
ax2.set_ylabel('satoshi', color='b')
ax2.grid(False)

ax3 = ax1.twinx()
ax3.plot(mdf.posix, np.sin(np.linspace(-5, 5, 288))*.3+.2, 'm-')
ax3.set_ylabel('performance', color='m')
ax3.grid(False)


# fig.tight_layout()
plt.show()
```
